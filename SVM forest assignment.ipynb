```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
```

```python
forest = pd.read_csv('forestfires.csv')
forest
```

```python
forest.drop(columns = {'month','day'},axis=1,inplace=True)
forest
```

```python
forest.describe()
```

```python
forest.isnull().sum()
```

```python
forest.size_category.value_counts()
```

```python
plt.figure(figsize=(10, 5))
sns.scatterplot(forest['temp'],forest['rain'],hue=forest['size_category']);
```

```python
d = {'small': 0, 'large': 1}
forest['size_category'] = forest['size_category'].map(d)
```

```python
forest.head()
```

```python
predictors = forest.iloc[:,0:28]
target = forest.iloc[:,28]
```

```python
def norm_func(i):
    x= (i-i.min())/(i.max()-i.min())
    return (x)
```

```python
from sklearn.svm import SVC
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test = train_test_split(predictors,target,test_size = 0.25, stratify = target)
```

```python
model_linear = SVC(kernel = "linear")
model_linear.fit(x_train,y_train)
pred_test_linear = model_linear.predict(x_test)

np.mean(pred_test_linear==y_test)
```

```python
model_poly = SVC(kernel = "poly")
model_poly.fit(x_train,y_train)
pred_test_poly = model_poly.predict(x_test)

np.mean(pred_test_poly==y_test)
```

```python
model_rbf = SVC(kernel = "rbf")
model_rbf.fit(x_train,y_train)
pred_test_rbf = model_rbf.predict(x_test)

np.mean(pred_test_rbf==y_test)
```

```python
model_sig = SVC(kernel = "sigmoid")
model_sig.fit(x_train,y_train)
pred_test_sig = model_sig.predict(x_test)

np.mean(pred_test_sig==y_test)
```
